# Variables in Python

Variables are used to store data values. In Python, you do not need to declare the data type explicitly.

Creating Variables

```python
name = "Ahmet"
age = 25
height = 1.80
is_student = True

print(name)
print(age)
print(height)
print(is_student)
```

Variable Types

```python
name = "John"      # String
age = 30           # Integer
salary = 5000.50   # Float
is_employed = True # Boolean

print(type(name))
print(type(age))
print(type(salary))
print(type(is_employed))
```
